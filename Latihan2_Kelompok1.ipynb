```python
import pandas as pd
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
url = 'https://raw.githubusercontent.com/gianiarda/latihan2/main/Iris.csv'
data = pd.read_csv(url)

print(data)
```

```python
data.sample(5)
```

```python
# split data into features (X) and labels (y)
X = data.iloc[:,1:5]
y = data.iloc[:,-1]
print(X.sample(5))
print(y.sample(5))
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white", color_codes=True)
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline

# see how many samples we have of each species 
data["Species"].value_counts()
```

# a. Petal Lenght vs Petal Width

```python
sns.FacetGrid(data, hue="Species", size=4) \
   .map(plt.scatter, "PetalLengthCm", "PetalWidthCm") \
   .add_legend();
```

```python
X_scaled = data.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics

inix=[]
iniy=[]

for i in range(2,11):
  inix.append(i)
  nclusters = i # this is the k in kmeans
  seed = 0

  km = KMeans(n_clusters=nclusters, random_state=seed)
  km.fit(X_scaled)

  # predict the cluster for each data point
  y_cluster_kmeans = km.predict(X_scaled)
  
  score = metrics.silhouette_score(X_scaled, y_cluster_kmeans)
  iniy.append(score)
  print("Silhouette score untuk k =",i, "adalah", score)

plt.plot(inix,iniy)
plt.xlabel ("Nilai k")
plt.ylabel ("Silhouette score")
plt.show
```

$k=2$ memberikan nilai Silhouette score tertinggi, sehingga nilai $k$ terbaik yang digunakan adalah $2$.

```python
from sklearn.mixture import GaussianMixture
from matplotlib.patches import Ellipse
def plot_cov_ellipse(cov, pos, col='b', nstd=2, ax=None, **kwargs):
    def eigsorted(cov):
        vals, vecs = np.linalg.eigh(cov)
        order = vals.argsort()[::-1]
        return vals[order], vecs[:,order]
    if ax is None:
        ax = plt.gca()
    vals, vecs = eigsorted(cov)
    theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))
    # Width and height are "full" widths, not radius
    width, height = 2 * nstd * np.sqrt(vals)
    ellip = Ellipse(xy=pos, width=width, height=height, angle=theta, color=col, alpha=0.2,**kwargs)
    ax.add_artist(ellip)
    return ellip
warna = ['navy', 'turquoise', 'darkorange','red','black','blue','green','yellow']
tanda = ['+','*','x','.','+','*','x','.']
for i in range (3,11):
  print(i,"cluster")
  estimators = dict((cov_type, GaussianMixture(n_components=i,
                   covariance_type=cov_type, max_iter=20, random_state=0))
                  for cov_type in ['spherical', 'diag', 'tied', 'full'])
  for (name,estimator) in estimators.items():
      estimator.fit(X_scaled) ## fits mixture model menggunakan EM
  for (name,gmm) in estimators.items():
      colors = warna[:i]
      markers = tanda[:i]
      for n, color in enumerate(colors):
          data = iris.data[iris.target == n]
          plt.scatter(data[:, 2], data[:, 3], marker=markers[n])
          if gmm.covariance_type == 'full':
              cov = gmm.covariances_[n]
          elif gmm.covariance_type == 'tied':
              cov = gmm.covariances_
          elif gmm.covariance_type == 'diag':
              cov = np.diag(gmm.covariances_[n])
          elif gmm.covariance_type == 'spherical':
              cov = np.eye(gmm.means_.shape[1]) * gmm.covariances_[n]
          #print(cov)
          pos = estimator.means_[n]
          #print(pos)
          plot_cov_ellipse(cov,pos,col=color)
      plt.title(name)
      plt.show()
```

```python
ar1=[]
ar2=[]
for i in range(2,11):
  GM = GaussianMixture(n_components=i, max_iter=20, random_state=0)
  GM.fit(X_scaled)
  ar1.append(i)
  ar2.append(metrics.silhouette_score(X_scaled, GM.predict(X_scaled)))
```

```python
plt.figure(figsize=(7,5))
plt.title("Silhouette Score K-means vs GMM pada Petal length vs width")
plt.plot(inix,iniy, label="K-means")
plt.plot(ar1, ar2, label='GMM')
plt.legend()
plt.xlabel ("n-component")
plt.ylabel ("Silhouette score")
plt.show
```
